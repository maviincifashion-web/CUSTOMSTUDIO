## Perform restore to HEAD

```python
import subprocess

repo_path = 'd:/IMRAN SOFTWARE DEVELOPER/MAIN APPLICATION/MAVIINCI/MaviinciApp'

print('Running git reset --hard HEAD...')
reset_result = subprocess.run(['git', 'reset', '--hard', 'HEAD'], cwd=repo_path, capture_output=True, text=True)
print('returncode:', reset_result.returncode)
print(reset_result.stdout)
print(reset_result.stderr)

print('Running git clean -fd...')
clean_result = subprocess.run(['git', 'clean', '-fd'], cwd=repo_path, capture_output=True, text=True)
print('returncode:', clean_result.returncode)
print(clean_result.stdout)
print(clean_result.stderr)
```
